```python
#Converting the columns to Numeric Dtypes as Required
df_train["Client_Income"].replace('$',np.nan,inplace=True)
df_train["Client_Income"]=pd.to_numeric(df_train["Client_Income"])
```

```python
df_train["Credit_Amount"].replace('$',np.nan,inplace=True)
df_train['Credit_Amount']=pd.to_numeric(df_train['Credit_Amount'])
```

```python
df_train["Loan_Annuity"].replace(["$", "#VALUE!"], np.nan, inplace = True)
df_train["Loan_Annuity"] = pd.to_numeric(df_train["Loan_Annuity"])
```

```python
df_train['Age_Days'].replace("x", np.nan, inplace = True)
df_train['Age_Days'] = pd.to_numeric(df_train['Age_Days'])
```

```python
df_train['Employed_Days'].replace("x", np.nan, inplace = True)
df_train['Employed_Days'] = pd.to_numeric(df_train['Employed_Days'])
```

```python
df_train['Registration_Days'].replace("x", np.nan, inplace = True)
df_train['Registration_Days'] = pd.to_numeric(df_train['Registration_Days'])
```

```python
df_train['ID_Days'].replace("x", np.nan, inplace = True)
df_train['ID_Days'] = pd.to_numeric(df_train['ID_Days'])
```

```python

df_train["Score_Source_3"].replace("&", np.nan, inplace=True)
df_train['Score_Source_3'] = pd.to_numeric(df_train['Score_Source_3'])
```

```python
# Now we have to calculate the mean to find whether Region Relative is under populated or over populated. To find that we have to calculate the mean
df_train["Population_Region_Relative"].replace(["#", "@"], np.nan, inplace=True)
df_train["Population_Region_Relative"] = pd.to_numeric(df_train["Population_Region_Relative"])

df_train.Population_Region_Relative = pd.to_numeric(df_train.Population_Region_Relative)
PRR_mean = df_train["Population_Region_Relative"].mean()
temp = df_train.copy()

temp.loc[temp['Population_Region_Relative'] < PRR_mean, 'Population_Region_Relative'] = 0  # for less populated areas
temp.loc[temp['Population_Region_Relative'] >= PRR_mean, 'Population_Region_Relative'] = 1  # for more populated areas
df_train['Population_Region_Relative'] = temp['Population_Region_Relative']
```

```python
# We are dropping the Customer ID
df_train.drop('ID',axis=1,inplace=True)
```

```python
from sklearn.impute import SimpleImputer

# Initialize SimpleImputer with mode strategy
si_mode = SimpleImputer(strategy='most_frequent')

# Fit and transform specific columns using correct indexing
# for Categorical Data we apply mode
df_train.iloc[:, 0:5] = si_mode.fit_transform(df_train.iloc[:, 0:5])
df_train.iloc[:, 8:16] = si_mode.fit_transform(df_train.iloc[:, 8:16])
df_train.iloc[:, 20:32] = si_mode.fit_transform(df_train.iloc[:, 20:32])
df_train.iloc[:, 37:38] = si_mode.fit_transform(df_train.iloc[:, 37:38])
```

```python
#for Numerical Data we apply Mean
si_mean=SimpleImputer(strategy="mean")
df_train.iloc[:, 6:8]=si_mean.fit_transform(df_train.iloc[:, 6:8])
df_train.iloc[:, 32:38]=si_mean.fit_transform(df_train.iloc[:, 32:38])
```

```python
#To Convert the decimal values into Integers
df_train.iloc[:, 5:6]=np.round(si_mean.fit_transform(df_train.iloc[:, 5:6]))
df_train.iloc[:, 16:20]=np.round(si_mean.fit_transform(df_train.iloc[:, 16:20]))
df_train.iloc[:, 36:37]=np.round(si_mean.fit_transform(df_train.iloc[:, 36:37]))
```

```python
#Lable Encoding
from sklearn.preprocessing import LabelEncoder

cols = ["Accompany_Client", "Client_Marital_Status", "Client_Income_Type", "Client_Education", "Client_Gender", "Loan_Contract_Type", "Client_Housing_Type", "Client_Occupation", "Client_Permanent_Match_Tag", "Client_Contact_Work_Tag", "Type_Organization"]
df_train[cols]=df_train[cols].apply(LabelEncoder().fit_transform)
```

```python
# Model Training
X=df_train.drop("Default",axis=1)
y=df_train.Default
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=27)

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf_classifier=RandomForestClassifier()
rf_classifier.fit(X_train,y_train)
predict=rf_classifier.predict(X_test)
accuracy_score(y_test,predict)*100
#  This is to check the accuracy score of the train_test_mode
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max.columns', None)
sns.set(style = "darkgrid", font_scale = 1.5)
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, StackingClassifier
from xgboost import XGBClassifier
#from catboost import CatBoostClassifier
#from lightgbm import LGBMClassifier


from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler
```

```python
# Precision of the model
training_score=[]
testing_score=[]
def model_prediction(model):
    model.fit(X_train,y_train)
    X_train_pred1=model.predict(X_train)
    X_test_pred1=model.predict(X_test)
    accuracy_score1=accuracy_score(y_train,X_train_pred1)*100
    accuracy_score2=accuracy_score(y_test,X_test_pred1)*100
    training_score.append(accuracy_score1)
    testing_score.append(accuracy_score2)
    print(f"Accuracy_Score of {model} model on Training Data is:",accuracy_score1)
    print(f"Accuracy_Score of {model} model on Testing Data is:",accuracy_score2)
    print("\n------------------------------------------------------------------------")
    print(f"Precision Score of {model} model is:",precision_score(y_test,X_test_pred1))
    print(f"Recall Score of {model} model is:",recall_score(y_test,X_test_pred1))
    print(f"F1 Score of {model} model is:",f1_score(y_test,X_test_pred1))
    print("\n------------------------------------------------------------------------")
    print(f"Confusion Matrix of {model} model is:")

    cm=confusion_matrix(y_test,X_test_pred1)
    plt.figure(figsize=(15,6))
    sns.heatmap(cm,cmap='viridis',annot=True,fmt='g')#fmt='g' stands for the floating numbers in heatmap
    plt.show()


```

```python
model_prediction(LogisticRegression())
```
